```python
import pandas as pd
import os
```

```python
# Specify the directory containing your data files
directory =  r'C:\Users\dhjrz\nlp\ionlab\AAPL'

# Initialize a dictionary to store filenames and sentences
data = {'filename': [], 'sentences': []}

# Iterate over files in the directory
for filename in os.listdir(directory):
    filepath = os.path.join(directory, filename)
    with open(filepath, 'r') as file:
        sentences = set(file.readlines())
        data['filename'].append(filename)
        data['sentences'].append(sentences)

# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
import json
preprocessed_data = []
for data_set in df['sentences']:
    combined_text = []
    for json_obj in data_set:
        item = json.loads(json_obj)
        if 'user_id_str' in item:
            del item['user_id_str']
        if 'created_at' in item:
            del item['created_at']
        combined_text.extend(item['text'])  # Extend the combined list
    preprocessed_data.append(combined_text)  # Append the combined list for each set

df['preprocessed_text'] = preprocessed_data
df.head(20)
```

```python
import nltk
from nltk.corpus import stopwords

# Download stopwords
nltk.download('stopwords')

# Define stopwords
nltk_stopwords = set(stopwords.words('english'))
additional_stopwords = {'rt', 'URL', 'AT_USER'}
all_stopwords = nltk_stopwords.union(additional_stopwords)
remove_characters = ['$']

def clean_tweet(words):
    cleaned_words = [
        word for word in words
        if word.lower() not in all_stopwords and not any(char in word for char in remove_characters)
    ]
    return cleaned_words

df['cleaned_tweet'] = df['preprocessed_text'].apply(clean_tweet)
```

```python
df = df.drop(columns=['sentences', 'preprocessed_text'])
df.head(20)
```

```python
# 날짜를 인덱스로 설정
aaplprice = pd.read_csv(r'C:\Users\dhjrz\nlp\ionlab\AAPL.csv')
aapl = aaplprice[(aaplprice['Date']>= '2014-01-01') & (aaplprice['Date']<='2015-12-31')]

df1 = pd.DataFrame()
df2 = pd.DataFrame()

df1['date'] = pd.to_datetime(aapl['Date'])
df2['date'] = pd.to_datetime(df['filename'])

df1.set_index('date', inplace=True)
df2.set_index('date', inplace=True)

# df1에는 있지만 df2에는 없는 날짜
missing_in_df2 = df1.index.difference(df2.index)
missing_in_df2
# missing = missing_in_df2.to_list()
```

```python
df['filename'] = pd.to_datetime(df['filename'], format='%Y-%m-%d', errors='coerce')

missing_df = pd.DataFrame({
    'filename': pd.to_datetime(missing_in_df2, format='%Y-%m-%d'),
    'cleaned_tweet': [[''] for _ in missing_in_df2]
})

# 기존 데이터프레임에 결측 데이터프레임을 추가
df = pd.concat([df, missing_df], ignore_index=True)

# 날짜 순으로 정렬하고 인덱스를 재설정
df = df.sort_values(by='filename').reset_index(drop=True)

# 결과 출력
df.head(20)
```

```python
import torch
from transformers import BertForSequenceClassification, BertTokenizer, pipeline
import pandas as pd

# Load pre-trained FinBERT model and tokenizer
finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone', num_labels=3)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
nlp = pipeline("text-classification", model=finbert, tokenizer=tokenizer, return_all_scores=True)

# Define sentiment analysis function using FinBERT
def get_sentiment_for_text(text):
    """Analyzes sentiment for a text string and returns sentiment scores."""
    text = str(text)  # Convert input to string
    # Truncate if needed
    tokens = tokenizer.tokenize(text)
    if len(tokens) > 510:
        tokens = tokens[:510]
    text = tokenizer.convert_tokens_to_string(tokens)

    result = nlp(text)

    # Debug: print the result to understand the output format
    print(result)

    # Extract probabilities directly using list comprehension
    positive_prob = next((item['score'] for item in result[0] if item['label'].lower() == 'positive'), 0)
    negative_prob = next((item['score'] for item in result[0] if item['label'].lower() == 'negative'), 0)
    neutral_prob = next((item['score'] for item in result[0] if item['label'].lower() == 'neutral'), 0)

    # Calculate sentiment score: -1 for negative, 1 for positive, and 0 for neutral
    sentiment_score = positive_prob - negative_prob

    sentiment_probs = {
        'label': max(result[0], key=lambda x: x['score'])['label'],
        'positive': positive_prob,
        'negative': negative_prob,
        'neutral': neutral_prob,
        'sentiment_score': sentiment_score
    }
    return sentiment_probs

# Apply sentiment analysis to each cleaned tweet in the DataFrame
df[['label', 'positive', 'negative', 'neutral', 'sentiment_score']] = df['cleaned_tweet'].apply(get_sentiment_for_text).apply(pd.Series)

    
print(df.head(20))
```
